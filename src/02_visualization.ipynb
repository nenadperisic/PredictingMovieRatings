```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("../data/train_final.csv")
test_data = pd.read_csv('../data/test_final.csv')
```

```python
all_data = pd.concat([train_data, test_data])
all_data = all_data.drop(columns = ['Genre', 'actors', 'director', 'companies', 'overview'])
# all_data.head()
```

```python
sns.distplot(all_data[all_data['horror'] == 1]['IMDB Score'])
sns.distplot(all_data[all_data['horror'] == 0]['IMDB Score'])
plt.legend(labels=['horror', 'not horror'])
```

```python
sns.distplot(all_data[all_data['documentary'] == 1]['IMDB Score'])
sns.distplot(all_data[all_data['documentary'] == 0]['IMDB Score'])
plt.legend(labels=['documentary', 'not documentary'])
```

```python
sns.distplot(all_data[all_data['budget'] >= 10000000]['IMDB Score'])
sns.distplot(all_data[all_data['budget'] < 10000000]['IMDB Score'])
plt.legend(labels=['more than ', 'less then '])
```

```python
sns.distplot(all_data[all_data['runtime'] >= 120]['IMDB Score'])
sns.distplot(all_data[all_data['runtime'] < 120]['IMDB Score'])
plt.legend(labels=['longer than 2 hours', 'less then 2 hours'])
```
